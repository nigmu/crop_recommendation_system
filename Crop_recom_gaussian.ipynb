```python
import pandas as pd

df = pd.read_csv("Crop_recommendation.csv")
df.head()
```

```python
x = df.drop(['label'], axis=1)
```

```python
from sklearn import preprocessing

Y = df['label']
encode = preprocessing.LabelEncoder()
y = encode.fit_transform(Y)
print("Label length: ",len(y))


label_encoder_mapping = dict(zip(encode.classes_, encode.transform(encode.classes_)))
print("Label and Encoded Value:")
for label, encoded_value in label_encoder_mapping.items():
    print(f"{label}: {encoded_value}")
```
